```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
# Separating the inputs and the outputs from the dataset
output_target = df['price_range']

input_data = df.drop('price_range',axis='columns')

```

```python
# since macine learning algorithims can only work with numbers, then 
# convert the words to numbers using encoders from sklearn

# importing labelEncoder from sklearn libary
# this code is not needed because we are working with
# a dataset in which all the data in each column are in numbers
# therefore we dont need to encode any labels

# but ust in case you have a dataset that requires label encoding, 
# you can run this celll, if you run you will get an error

from sklearn.preprocessing import LabelEncoder

def LabelEncod(input_data):
    # the input_data must be a pandas dataframe 
     
    for i in input_data:
        try:
            tempEncode = LabelEncoder()
            input_data['Encoded'+str(i)] = tempEncode.fit_transform(input_data[i])
        except e:
            continue
        
        
    return 


```

```python
# USING THE DECISION TREE MACHINE LEARNING ALGORITHIM FROM SKLEARN

from sklearn import tree
# CREATING A MODEL TO E TRAINED
PhonePrice_model = tree.DecisionTreeClassifier()
```

```python
# TRAINING THE MODEL USING THE TRAIN DATA SET
PhonePrice_model.fit(input_data, output_target)
```

```python
# checking to see if the score will be 1 for the same dataset
PhonePrice_model.score(input_data, output_target)
```

```python
# now prediction the price for the phones in the test dataset


# getting the data
df_test = pd.read_csv("test .csv") # noting that the test.csv file naming has a space like "test .csv"
```

```python
# filtering the dataaset and removing the columns that are not needed for training the dataset

 
# Remove column name 'id
df_test = df_test.drop(['id'], axis=1)
```

```python
# predicting for the test dataset
test_output = PhonePrice_model.predict(df_test)

```
